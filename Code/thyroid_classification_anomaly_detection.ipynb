```python
import numpy as np
import pandas as pd
import math
```

```python
df= pd.read_csv("./thyroid0387.data",header=None, low_memory=False)
df.head()
```

```python
X = df.drop([5,29,28,17,19,23,25,27],axis=1)
```

```python
X_new=X.replace('?',np.nan)
X_new
```

```python
X_new.shape
```

```python
X_new2= X_new.dropna()
sick=df.iloc[:,5]
X_new2.columns = X_new2.columns.astype("str")
X_new3 =X_new2.rename(columns={"0":"AGE","1":"SEX","2":"ON THYROXINE","3":"QUERY ON THYROXINE",
                              "4":"ON ANTITHYROID MEDICATION","6":"PREGNENT","7":"THYROID SURGERY",
                              "8":"I131 TREATMENT","9": "QUERY HYPOTHYROID","10":"QUERY HYPERTHYROID",
                              "11":"LITHIUM","12":"GOITRE","13":"TUMOR","14":"HYPOPITUITARY","15":"PSYCH",
                              "16":"TSH MEASURED","18":"T3 MEASURED","20":"TT4 MEASURED","21":"TT4",
                              "22":"T4U","24":"FTI MEASURED","26":"TBG MEASURED"})
X_new3['TT4'] = pd.to_numeric(X_new3['TT4'])
X_new4= pd.get_dummies(X_new3)
X_new4['sick'] = sick
def convert(x):
    if x == 'f':
        return 0
    elif x == 't':
        return 1
X_new4['sick_new']= X_new4['sick'].apply(lambda x : convert(x))
X_new4['sick_new'].unique()
X_new5= X_new4.drop('sick',axis=1)
df_thyroid_preds= X_new5[['AGE','TT4','SEX_F','SEX_M','ON THYROXINE_f','ON THYROXINE_t','T3 MEASURED_f','T3 MEASURED_t','FTI MEASURED_f',
                     'FTI MEASURED_t','sick_new']]
X = df_thyroid_preds.drop('sick_new', axis =1)
y = df_thyroid_preds['sick_new']
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, cross_validate, train_test_split
from sklearn.utils import shuffle
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
X = scaler.fit_transform(X)
y = y.reset_index()['sick_new']
```

```python
seed=42
X, y = shuffle(X, y, random_state=seed)
```

```python
## Splitting X and Y into train and test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
## Number of training articles and test articles

print("Number of training samples:", X_train.shape[0])
print("Number of test samples:", X_test.shape[0])
```

```python
## Number of samples by class

print("Number of training samples for class 0: {}".format(y_train.value_counts()[0]))
print("Number of training samples for class 1: {}".format(y_train.value_counts()[1]))
```

```python
## Extracting training samples corresponding to class 0

X_train_zeros = X_train[y_train == 0]
```

```python
## training One Class SVM (Novelty Detection Model) to predict the outlier class (class 1 for this example)

from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
from sklearn.metrics import confusion_matrix

## Kernel:RBF
trainSVM = OneClassSVM(kernel='rbf', nu=0.9, gamma='auto').fit(X_train_zeros)
y_test_predict = trainSVM.predict(X_test)
y_test_predict[y_test_predict==1] = 0
y_test_predict[y_test_predict==-1] = 1

## Calculating Confusion matrix, recall and precision for the outlier class 1
cf_matrix = confusion_matrix(y_test, y_test_predict)
print("Confusion Matrix:", cf_matrix)
recall = (cf_matrix[1][1]/np.float(cf_matrix[1][1] + cf_matrix[1][0]))
precision = (cf_matrix[1][1]/np.float(cf_matrix[1][1] + cf_matrix[0][1]))
print("Recall:", recall)
print("Precision:", precision)
print("F1 score:", (2*recall*precision/(recall + precision)))
```

```python
## Kernel:Polynomial
trainSVM = OneClassSVM(kernel='poly', degree=3, nu=0.9, gamma='auto').fit(X_train_zeros)
y_test_predict = trainSVM.predict(X_test)
y_test_predict[y_test_predict==1] = 0
y_test_predict[y_test_predict==-1] = 1

## Calculating Confusion matrix, recall and precision for the outlier class 1
cf_matrix = confusion_matrix(y_test, y_test_predict)
print("Confusion Matrix:", cf_matrix)
recall = (cf_matrix[1][1]/np.float(cf_matrix[1][1] + cf_matrix[1][0]))
precision = (cf_matrix[1][1]/np.float(cf_matrix[1][1] + cf_matrix[0][1]))
print("Recall:", recall)
print("Precision:", precision)
print("F1 score:", (2*recall*precision/(recall + precision)))
```
