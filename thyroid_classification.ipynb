```python
X_new4= pd.get_dummies(X_new3)
```

```python
X_new4['sick'] = sick
```

```python
def convert(x):
    if x == 'f':
        return 0
    elif x == 't':
        return 1
X_new4['sick_new']= X_new4['sick'].apply(lambda x : convert(x))
```

```python
X_new4['sick_new'].unique()
```

```python
X_new4.info()
```

```python
X_new5= X_new4.drop('sick',axis=1)
```

```python
X_new5.head()
```

```python
df_thyroid_preds= X_new5[['AGE','TT4','SEX_F','SEX_M','ON THYROXINE_f','ON THYROXINE_t','T3 MEASURED_f','T3 MEASURED_t','FTI MEASURED_f',
                     'FTI MEASURED_t','sick_new']]
```

```python
df_thyroid_preds.head()
```

```python
X = df_thyroid_preds.drop('sick_new', axis =1)
```

```python
y = df_thyroid_preds['sick_new']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= 0.2,random_state= 101)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
 prediction= dtree.predict(X_test) 
```

```python
df_thyroid_preds_new_dt = pd.DataFrame({'Actual': y_test, 'Predicted': prediction})
```

```python
df_thyroid_preds_new_dt
```

```python
from sklearn import metrics
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, prediction))
```

## RANDOM FOREST

```python
from sklearn .ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
df_thyroid_preds_new_rf = pd.DataFrame({'Actual': y_test, 'Predicted_rf': rfc_pred})
```

```python
df_thyroid_preds_new_rf
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, rfc_pred))
```
